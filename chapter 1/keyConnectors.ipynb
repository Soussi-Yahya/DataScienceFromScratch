```python
from collections import Counter

def not_the_same(user,other_user) : 
    #two users are not the same if their ids are different 
    return user["id"] != other_user["id"]

def not_friends(user , other_user) : 
    #returns True if other_user is NOT already a friend of user
    #It checks every friend in user["friends"] and ensures that none of them are the same as other_user
    return all(not_the_same(friend , other_user) for friend in user["friends"])

def friends_of_friend_ids(user) : 
    #Loop over all of the userâ€™s friends
    #Loop over all friends of each friend
    #Filter out invalid FOAFs(They are not the user ,They are not already a friend )
    #and finally Extract their IDs and count them
    return Counter(foaf["id"]
                   for friend in user["friends"]
                   for foaf in friend["friends"]
                   if not_the_same(user,foaf)
                   and not_friends(user,foaf))

```

```python
from collections import defaultdict

#keys are interests , values are lists of user_ids with that interest
user_ids_by_interest = defaultdict(list)

for user_id , interest in interests : 
    user_ids_by_interest[interest].append(user_id)

#keys are user_ids , values are lists of interests for that user_id
interests_by_user_id = defaultdict(list)

for user_id, interest in interests:
    interests_by_user_id[user_id].append(interest)
```

```python
def most_common_interests_with(user) : 
    #iterate over the user's interests 
    #for each interest , iterate over the other users with that interest 
    #keep count of how many times we see each other user
    return Counter(interested_user_id
                   for interest in interests_by_user_id[user["id"]]
                   for interested_user_id in user_ids_by_interest[interest]
                   if interested_user_id != user["id"])
```
